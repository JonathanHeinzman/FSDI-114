```python
# ======================================
# Single Linked List Implementation
# ======================================

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        # Adds a new node at the end of the list
        new_node = Node(value)

        # If the list is empty, new node becomes the head
        if self.head is None:
            self.head = new_node
            return
        
        # Otherwise, walk to the end and attach it
        current = self.head
        while current.next is not None:
            current = current.next

        current.next = new_node

    # Exercise 2 - print_list(). Print all the values in the list
    def print_list(self):
        current = self.head

        if current is None:
            print("List is empty!")
            return
        
        while current is not None:
            print(current.value)
            current = current.next
    
    
    # Exercise 3 - Count nodes
    def count_nodes(self):
        count = 0
        current = self.head
        while current is not None:
            count += 1
            current = current.next
        return count
        
    # Exercise 4 - 
    def prepend(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node


    # Exercise 5 - Remove
    def remove(self, value):
        if self.head is None:
            return
        
        if self.head.value == value:
            self.head = self.head.next
            return
        
        current = self.head
        while current.next is not None:
            if current.next.value == value:
                current.next = current.next.next
                return
            current = current.next


        
# Exercise 1: Create a linked list and append values
myList = LinkedList()
myList.append(10)
myList.append(20)
myList.append(30)


print("Initial list (before append)")
myList.print_list()
print("Node count: ", myList.count_nodes())

myList.prepend(50)
print("List (after append)")
myList.print_list()

myList.remove(30)
print("List (after remove)")
myList.print_list()
```
