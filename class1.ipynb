Challenge 2

START

If the list is empty:
    Display "List is empty"
    STOP

Set larges to the first value in the list

For each number in the list starting from the second item:
    If the number is greater than largest:
        Update largest to this number

Display largest

END

```python
nums = [12, 4, 27, 3, 211, 50, 7]

def find_largest(nums):
    if not nums:
        return "No values"
    
    largest = nums[0]

    for n in nums[1:]:
        if n > largest:
            largest = n
    
    return largest

print(find_largest(nums))
```
