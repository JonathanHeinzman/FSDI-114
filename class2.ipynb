# Stacks:

A stack is an ordered collection of elements where items are added and removed from one end, commonly referred to as the "top". The ordering principle is referred to as LIFO (Last In, First Out).

```python
# Challenge 1

stack = []

stack.append("A")
stack.append("B")
stack.append("C")

print(stack) # Stack: ['A','B','C']

top = stack.pop()

print(f"Stack: {stack} | Last value: {top}")
print(f"Next Value to be removed: {stack.pop()}")

stack.append("D")
stack.pop()
print(f"{stack}")
```

```python
# Challenge 4

word = "HELLO"
stack = []

for letter in word:
    stack.append(letter)

reversed_word = ""
while len(stack) > 0:
    reversed_word += stack.pop()

print(reversed_word)
```

# Queues:

A queue is an ordered collection of elementes where items are added and removed from **opposite** ends, commonly referred to as the "front" and "back" or "rear" of the queue.  The ordering principle is referred to as FIFO (First In, First Out) or "first come first serverd".

```python
# Challenge 5

queue = []

queue.append("10")
queue.append("20")
queue.append("30")

print(queue)
```
