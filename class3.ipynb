```python
# Assignment 3 Stack
class Stack:
# Challenge 1
    def __init__(self):
        self.items = []

# Challenge 2 - Check if it is empty
    def is_empty(self):
        return len(self.items) == 0

# Challenge 3 - Push a new item
    def push(self, item):
        self.items.append(item)

# Challenge 4 - Pop an item
    def pop(self):
        if self.is_empty():
            return None
        return self.items.pop()
    
# Challenge 5 - Peek at the top item & Return the size
    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]
    
    def size(self):
        return len(self.items)
```

```python
s = Stack()

s.push(5)
s.push(10)
s.push(15)

print("Stack peek: ", s.peek())
print("Stack pop: ", s.pop())
print("Stack size: ", s.size())
```

```python
# Challenge 6 Queue
class Queue:
    
    def __init__(self):
        self.items = []

# Challenge 7 - Check if the queue is empty
    def is_empty(self):
        return len(self.items) == 0

# Challenge 8 - Enqueue
    def enqueue(self, item):
        self.items.append(item)

# Challenge 9 - Dequeue
    def dequeue(self):
        if self.is_empty():
            return None
        return self.items.pop(0)

# Challenge 10 - Return the size of the queue
    def size(self):
        return len(self.items)
```

```python
q = Queue()

q.enqueue("A")
q.enqueue("B")
q.enqueue("C")

print("Queue dequeue: ", q.dequeue())
print("Queue size: ", q.size())
```
